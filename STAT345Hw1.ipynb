```python
import statistics

original_list = [15, 18, 10, 17, 14, 16, 35, 28, 22, 17, 23, 12, 15, 28, 10, 20]

print("The original mean is", statistics.mean(original_list))
print("The original median is", statistics.median(original_list))
print("The original standard deviation is", statistics.stdev(original_list))

#add 5 to original list
new_list = []
for num in original_list: 
    new_num = num + 5
    new_list.append(new_num)
    
print("The new mean is", statistics.mean(new_list))
print("The new median is", statistics.median(new_list))
print("The new standard deviation is", statistics.stdev(new_list))
```
